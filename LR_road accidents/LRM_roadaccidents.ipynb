#capping - change the outlier values to upper (or) lower limit values.

```python
upper_limit= data['Accident_severity'].mean() + 3*data['Accident_severity'].std()
lower_limit= data['Accident_severity'].mean() - 3*data['Accident_severity'].std()
print('upper limit;', upper_limit)
print('lower limit;',lower_limit)
```

```python
#find the ouliers
data.loc[(data['Accident_severity']> upper_limit) | (data['Accident_severity']<lower_limit)]
```

```python
#trimming - delete the outlier data
new_data = data.loc[(data['Accident_severity']< upper_limit) | (data['Accident_severity']>lower_limit)]
print('old data:', len(data))
print('new data:',len(new_data))
```
